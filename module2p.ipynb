```python
# import statements
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('companies.csv')
```

```python
df.head()
```

```python
# Question 1:
# Using matplotlib/seaborn, create a countplot that shows a count of how many of the companies in the data set are from each state. 

# sorted graph
sorted_data = df['state'].value_counts().sort_values(ascending = False)

plt.figure(figsize = (10,6))
sns.countplot(x='state', data=df, order = sorted_data.index)
plt.title('Company Count by State')
plt.xticks(rotation = 90)
plt.show()
```

```python
# Question 3:
# Using matplotlib/seaborn, create a bar chart that shows a sum of how many total job openings there are in each industry.

sns.barplot(x='Industry', y='job_openings', data=df, estimator = 'sum', errorbar = None)
plt.xticks(rotation = 90)
plt.title('Job Openings by Industry')
plt.show()
```

```python
# Question 5:
# Using matplotlib/seaborn, create a line chart that shows the average growth rate for each founded year.

# Create a new dataframe
df_rate = df.groupby('founded')['growth_percentage'].mean().reset_index()

# Graph
plt.title('Average Growth by Year Founded')
sns.lineplot(x='founded', y='growth_percentage', data=df_rate)
plt.show()
```

```python
# Question 7
# Using matplotlib/seaborn, create a scatter plot that compares the number of employees to estimated revenues for each company.

sns.scatterplot(x='employees', y='estimated_revenues', data=df)
plt.title('Estimated Revenue by Total Employees')
plt.ylabel('Estimated Revenue in Millions')
plt.xlabel('Total Employees')
plt.show()
```

```python
# Question 9
# Using matplotlib/seaborn, create a histogram that measures the number of companies founded in each year.

sns.histplot(x='founded', data=df, bins = 50)
plt.title('Companies Founded Each Year')
plt.show()
```
